```python
sns.histplot(df['Age'])
plt.title('Age Distribution')
plt.show()
```

```python
sns.histplot(df['Gender'])
plt.title('Gender Distribution')
plt.show()
```

```python
correlation_matrix = df.select_dtypes(include=[np.number]).corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', hue='Heart_Attack_Outcome', data=df)
plt.title('Heart Attack Outcome by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
bins = [20, 30, 40, 50, 60, 70, 80]
labels = ['20-29', '30-39', '40-49', '50-59', '60-69', '70-79']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

plt.figure(figsize=(10, 6))
sns.countplot(x='Age_Group', hue='Heart_Attack_Outcome', data=df)
plt.title('Heart Attack Outcome by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
pclass_counts = df["Alcohol_Intake"].value_counts()
print(pclass_counts)

plt.pie(pclass_counts, labels=["Moderate", "Low", "High"],
        autopct="%1.3f%%", colors=["gold", "silver", "brown"])
plt.show()
```

```python
pclass_counts = df["Smoking_Status"].value_counts()
print(pclass_counts)

plt.pie(pclass_counts, labels=["No", "Yes"],
        autopct="%1.3f%%", colors=["green", "red"])
plt.show()
```

```python
pclass_counts = df["Physical_Activity"].value_counts()
print(pclass_counts)

plt.pie(pclass_counts, labels=["Sedentary", "Active", "Highly_Active"],
        autopct="%1.3f%%", colors=["brown", "silver", "gold"])
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Physical_Activity', hue='Heart_Attack_Outcome', data=df)
plt.title('Heart Attack Outcome by Gender')
plt.xlabel('Physical_Activity')
plt.ylabel('Count')
plt.show()
```
